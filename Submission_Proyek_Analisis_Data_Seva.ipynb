## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa saja produk yang paling banyak di beli dan yang paling sedikit di beli?

```python
max_products = product_name_df["product_category_name"].value_counts().head(5)
max_products_df = pd.DataFrame({
    'Nama Produk': max_products.index,
    'Total': max_products.values
})
max_products_df
```

```python
min_products = product_name_df["product_category_name"].value_counts().tail(5)
min_products_df = pd.DataFrame({
    'Nama Produk': min_products.index,
    'Total': min_products.values
})
min_products_df
```

```python
# Membuat visualisasi dengan horizontal bar chart untuk produk terlaris
plt.figure(figsize=(10, 6))

# Menentukan produk terlaris
top_product = max_products.idxmax()

# Memflip urutan data agar yang terlaris berada di atas
max_products_sorted = max_products.sort_values()

# Menggambar horizontal bar chart untuk produk terlaris
max_products_sorted.plot(kind='barh', color=['skyblue' if prod != top_product else 'navy' for prod in max_products_sorted.index])

# Menambahkan label pada bar
for i, val in enumerate(max_products_sorted.values):
    plt.text(val, i, f'{int(val)}', ha='left', va='center')

plt.title('Produk Terlaris')
plt.xlabel('Jumlah Pembelian')
plt.ylabel('Nama Produk')

plt.tight_layout()
plt.show()
```

```python
# Membuat visualisasi dengan horizontal bar chart untuk produk terlaris
plt.figure(figsize=(10, 6))

# Menentukan produk terlaris
top_product = min_products.idxmin()

# Menggambar horizontal bar chart untuk produk terlaris
min_products.plot(kind='barh', color=['navy' if prod == top_product else 'skyblue' for prod in min_products.index])

# Menambahkan label pada bar
for i, val in enumerate(min_products.values):
    plt.text(val, i, f'{int(val)}', ha='left', va='center')

plt.title('Produk Paling tidak Laris')
plt.xlabel('Jumlah Pembelian')
plt.ylabel('Nama Produk')

plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana tingkat kepuasan pelanggan terhadap layanan dan produk yang ditawarkan?

```python
# Definisikan ukuran figur
plt.figure(figsize=(10, 8))

# Buat histogram
plt.hist(order_reviews_df['review_score'], bins=5, color='skyblue', edgecolor='black')

# Judul
plt.title('Histogram Skor Ulasan', fontsize=16, fontweight='bold')

# Label sumbu x dan y
plt.xlabel('Skor Ulasan', fontsize=12)
plt.ylabel('Jumlah Ulasan', fontsize=12)

# Tambahkan grid untuk memperjelas pembacaan
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hapus spines (garis pinggir) atas dan kanan
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Tampilkan visualisasi
plt.show()
```

### Pertanyaan 3: Kota bagian mana yang paling tinggi tingkat pembeliannya?


```python
# Menghitung jumlah pembeli per negara bagian
jumlah_pembeli_per_state = geolocation_df['geolocation_state'].value_counts()

# Plot bar untuk jumlah pembeli per negara bagian
plt.figure(figsize=(12, 8))
jumlah_pembeli_per_state.plot(kind='bar', color='skyblue')

# Judul
plt.title('Jumlah Pembeli per Kota', fontsize=16, fontweight='bold')

# Label sumbu x dan y
plt.xlabel('Kota', fontsize=12)
plt.ylabel('Jumlah Pembeli', fontsize=12)

# Rotasi label sumbu x
plt.xticks(rotation=45, ha='right', fontsize=10)

# Tambahkan grid untuk memperjelas pembacaan
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hapus spines (garis pinggir) atas dan kanan
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah pembeli per negara bagian
jumlah_pembeli_per_state = geolocation_df['geolocation_state'].value_counts()

# Mengambil 10 negara bagian dengan jumlah pembeli terbanyak
top_10_states = jumlah_pembeli_per_state.head(10)

# Visualisasi bar plot jumlah pembeli per negara bagian (hanya 10 teratas)
plt.figure(figsize=(12, 8))
top_10_states.plot(kind='bar', color='skyblue')

# Judul
plt.title('Top 10 Kota dengan Jumlah Pembeli Tertinggi', fontsize=16, fontweight='bold')

# Label sumbu x dan y
plt.xlabel('Kota', fontsize=12)
plt.ylabel('Jumlah Pembeli', fontsize=12)

# Rotasi label sumbu x
plt.xticks(rotation=45, ha='right', fontsize=10)

# Tambahkan grid untuk memperjelas pembacaan
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hapus spines (garis pinggir) atas dan kanan
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah pembeli per negara bagian
jumlah_pembeli_per_state = geolocation_df['geolocation_state'].value_counts()

# Mengambil 10 negara bagian dengan jumlah pembeli terbanyak
top_10_states = jumlah_pembeli_per_state.head(10)

# Visualisasi pie chart proporsi jumlah pembeli per negara bagian (hanya 10 teratas)
plt.figure(figsize=(8, 8))
plt.pie(top_10_states, labels=top_10_states.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Proporsi Jumlah Pembeli per Kota (Top 10)', fontsize=16, fontweight='bold')
plt.axis('equal')  # Mengatur aspek rasio agar lingkaran terlihat bulat
plt.show()
```
